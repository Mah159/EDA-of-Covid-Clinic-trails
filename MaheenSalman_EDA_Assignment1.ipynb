On the given dataset perform the following task:
1. Explore the dataset by checking shape, columns, and see the first/last 'n' rows using
head/tail. (n= 5,15,30)

```python
# Importing necessary libraries
import pandas as pd

# Loading the dataset
df = pd.read_csv('COVID clinical trials.csv')

# Checking the shape of the dataset
shape = df.shape

# Checking the columns of the dataset
columns = df.columns.tolist()

# Viewing the first 'n' rows
head_5 = df.head(5)
head_15 = df.head(15)
head_30 = df.head(30)

# Viewing the last 'n' rows
tail_5 = df.tail(5)
tail_15 = df.tail(15)
tail_30 = df.tail(30)

shape, columns, head_5, head_15, head_30, tail_5, tail_15, tail_30
```

Extract and document key insights by exploring column data types and applying info and describe

```python
# Checking the data types of the columns
info = df.info()

# Generating descriptive statistics
describe = df.describe()

info, describe
```

Determine the missing values in the whole dataset and analyze missing values in each column.

```python
# Checking for missing values in the whole dataset
total_missing_values = df.isnull().sum().sum()

# Analyzing missing values in each column
missing_values_per_column = df.isnull().sum()

total_missing_values, missing_values_per_column
```

Calculate the sum of duplicate rows

```python
# Calculating the sum of duplicate rows
sum_of_duplicate_rows = df.duplicated().sum()

sum_of_duplicate_rows
```

Make 5 questions and solve them by using group by.

```python
# Answering question 1: How many clinical trials are there for each phase?
trials_per_phase = df.groupby('Phases')['NCT Number'].count()

# Answering question 2: What is the average enrollment for each phase?
average_enrollment_per_phase = df.groupby('Phases')['Enrollment'].mean()

# Answering question 3: How many clinical trials are there for each intervention type?
trials_per_intervention_type = df.groupby('Interventions')['NCT Number'].count()

# Answering question 4: What is the average enrollment for each intervention type?
average_enrollment_per_intervention_type = df.groupby('Interventions')['Enrollment'].mean()

# Answering question 5: How many clinical trials are there for each country?
trials_per_country = df.groupby('Locations')['NCT Number'].count()

trials_per_phase, average_enrollment_per_phase, trials_per_intervention_type, average_enrollment_per_intervention_type, trials_per_country
```

Make 5 questions and solve them by using conditional statements

```python
# Answering question 1: How many clinical trials are in Phase 3?
trials_in_phase_3 = df[df['Phases'] == 'Phase 3'].shape[0]

# Answering question 2: What is the average enrollment for clinical trials in Phase 3?
average_enrollment_phase_3 = df[df['Phases'] == 'Phase 3']['Enrollment'].mean()

# Answering question 3: How many clinical trials use the intervention type 'Drug'?
trials_with_drug_intervention = df[df['Interventions'] == 'Drug'].shape[0]

# Answering question 4: What is the average enrollment for clinical trials that use the intervention type 'Drug'?
average_enrollment_drug_intervention = df[df['Interventions'] == 'Drug']['Enrollment'].mean()

# Answering question 5: How many clinical trials are there in the United States?
trials_in_us = df[df['Locations'] == 'United States'].shape[0]

trials_in_phase_3, average_enrollment_phase_3, trials_with_drug_intervention, average_enrollment_drug_intervention, trials_in_us
```
